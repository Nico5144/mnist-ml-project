```python
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt

# 1. Load the MNIST dataset from Keras

(X_train, y_train), (X_test, y_test) = mnist.load_data()

# 2. Pick 25 random images from the training set

idxs = np.random.choice(len(X_train), 25, replace=False)

# 3. Display them in a 5×5 grid

fig, axes = plt.subplots(5, 5, figsize=(6,6))
for ax, i in zip(axes.flatten(), idxs):
    ax.imshow(X_train[i], cmap='gray')  # render as grayscale
    ax.axis('off')                       # hide axes ticks
plt.suptitle('25 Random MNIST Samples')
plt.show()
```

```python
# Calculate the overall mean and standard deviation of pixel intensities

mean_pixel = X_train.mean()
std_pixel  = X_train.std()
print(f"Mean pixel value: {mean_pixel:.2f}")
print(f"Std pixel value: {std_pixel:.2f}")
```

```python
import matplotlib.pyplot as plt

# Count how many examples of each digit we have

counts = np.bincount(y_train)   # bincount returns count of each integer label
labels = np.arange(10)

# Plot a bar chart

plt.figure(figsize=(6,4))
plt.bar(labels, counts)
plt.xlabel('Digit')
plt.ylabel('Frequency')
plt.title('MNIST Class Distribution (Training Set)')
plt.xticks(labels)
plt.show()
```

```python
import pandas as pd

distribution_df = pd.DataFrame({
    'Digit': labels,
    'Count': counts
})
distribution_df
```
